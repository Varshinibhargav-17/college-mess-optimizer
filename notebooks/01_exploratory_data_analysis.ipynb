```python
# ============================================
# CELL 6: TEMPORAL PATTERNS (Day of Week)
# ============================================

print("="*60)
print("PATTERN: Day of Week Analysis")
print("="*60)

# Average attendance by day
day_attendance = df.groupby('day_of_week')['students_attended'].agg(['mean', 'std', 'count'])
print("\nAverage Attendance by Day:")
print(day_attendance)

# Visualization
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_avg = df.groupby('day_of_week')['students_attended'].mean().reindex(day_order)
day_avg.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Attendance by Day of Week', fontsize=14, fontweight='bold')
plt.xlabel('Day of Week')
plt.ylabel('Average Students')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)

plt.subplot(1, 2, 2)
sns.boxplot(data=df, x='day_of_week', y='students_attended', order=day_order)
plt.title('Attendance Distribution by Day', fontsize=14, fontweight='bold')
plt.xlabel('Day of Week')
plt.ylabel('Students Attended')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# Insights
max_day = day_avg.idxmax()
min_day = day_avg.idxmin()
print(f"\nüí° INSIGHT:")
print(f"   - Highest attendance: {max_day} ({day_avg[max_day]:.0f} students)")
print(f"   - Lowest attendance: {min_day} ({day_avg[min_day]:.0f} students)")
print(f"   - Difference: {day_avg[max_day] - day_avg[min_day]:.0f} students")
```

```python
# ============================================
# CELL 7: MEAL TYPE ANALYSIS
# ============================================

print("="*60)
print("PATTERN: Meal Type Analysis")
print("="*60)

meal_stats = df.groupby('meal_type')['students_attended'].agg(['mean', 'std', 'min', 'max'])
print("\nAttendance by Meal Type:")
print(meal_stats)

plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
meal_avg = df.groupby('meal_type')['students_attended'].mean()
meal_avg.plot(kind='bar', color=['#FF6B6B', '#4ECDC4', '#45B7D1'], edgecolor='black')
plt.title('Average Attendance by Meal Type', fontsize=14, fontweight='bold')
plt.xlabel('Meal Type')
plt.ylabel('Average Students')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)

plt.subplot(1, 2, 2)
sns.violinplot(data=df, x='meal_type', y='students_attended')
plt.title('Attendance Distribution by Meal Type', fontsize=14, fontweight='bold')
plt.xlabel('Meal Type')
plt.ylabel('Students Attended')

plt.tight_layout()
plt.show()

print(f"\nüí° INSIGHT:")
print(f"   - Most popular: {meal_avg.idxmax()} ({meal_avg.max():.0f} students)")
print(f"   - Least popular: {meal_avg.idxmin()} ({meal_avg.min():.0f} students)")
```

```python
# ============================================
# CELL 8: MENU POPULARITY ANALYSIS
# ============================================

print("="*60)
print("PATTERN: Menu Item Popularity")
print("="*60)

menu_stats = df.groupby('menu_item').agg({
    'students_attended': ['mean', 'count'],
    'menu_popularity': 'first'
}).round(2)

menu_stats.columns = ['Avg Attendance', 'Times Served', 'Popularity Score']
menu_stats = menu_stats.sort_values('Avg Attendance', ascending=False)

print("\nTop 10 Most Popular Menu Items:")
print(menu_stats.head(10))

print("\nBottom 5 Least Popular Menu Items:")
print(menu_stats.tail(5))

# Visualization
plt.figure(figsize=(14, 8))

plt.subplot(2, 1, 1)
top_items = menu_stats.head(10)
plt.barh(range(len(top_items)), top_items['Avg Attendance'], color='green', alpha=0.7)
plt.yticks(range(len(top_items)), top_items.index)
plt.xlabel('Average Students Attended')
plt.title('Top 10 Most Popular Menu Items', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()
plt.grid(axis='x', alpha=0.3)

plt.subplot(2, 1, 2)
bottom_items = menu_stats.tail(10)
plt.barh(range(len(bottom_items)), bottom_items['Avg Attendance'], color='red', alpha=0.7)
plt.yticks(range(len(bottom_items)), bottom_items.index)
plt.xlabel('Average Students Attended')
plt.title('Bottom 10 Least Popular Menu Items', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()
plt.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\nüí° INSIGHT:")
print(f"   - Most popular item: {menu_stats.index[0]} ({menu_stats.iloc[0]['Avg Attendance']:.0f} students)")
print(f"   - Least popular item: {menu_stats.index[-1]} ({menu_stats.iloc[-1]['Avg Attendance']:.0f} students)")
```

```python
# ============================================
# CELL 9: WEATHER IMPACT
# ============================================

print("="*60)
print("PATTERN: Weather Impact on Attendance")
print("="*60)

weather_stats = df.groupby('weather')['students_attended'].agg(['mean', 'std', 'count'])
print("\nAttendance by Weather:")
print(weather_stats)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
weather_avg = df.groupby('weather')['students_attended'].mean()
weather_avg.plot(kind='bar', color=['#FFD700', '#87CEEB', '#778899'], edgecolor='black')
plt.title('Average Attendance by Weather', fontsize=14, fontweight='bold')
plt.xlabel('Weather')
plt.ylabel('Average Students')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)

plt.subplot(1, 2, 2)
sns.boxplot(data=df, x='weather', y='students_attended')
plt.title('Attendance Distribution by Weather', fontsize=14, fontweight='bold')
plt.xlabel('Weather')
plt.ylabel('Students Attended')

plt.tight_layout()
plt.show()

print(f"\nüí° INSIGHT:")
rainy_avg = weather_stats.loc['Rainy', 'mean']
sunny_avg = weather_stats.loc['Sunny', 'mean']
diff = rainy_avg - sunny_avg
print(f"   - Rainy day attendance: {rainy_avg:.0f} students")
print(f"   - Sunny day attendance: {sunny_avg:.0f} students")
print(f"   - Difference: {diff:+.0f} students ({(diff/sunny_avg*100):+.1f}%)")
```

```python
# ============================================
# CELL 10: SPECIAL CONDITIONS IMPACT
# ============================================

print("="*60)
print("PATTERN: Special Conditions Impact")
print("="*60)

conditions = ['is_exam_period', 'is_festival', 'is_weekend', 'is_start_semester']

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for idx, condition in enumerate(conditions):
    condition_stats = df.groupby(condition)['students_attended'].mean()
    
    ax = axes[idx]
    condition_stats.plot(kind='bar', ax=ax, color=['#E74C3C', '#2ECC71'], edgecolor='black')
    ax.set_title(f'Attendance: {condition.replace("_", " ").title()}', fontsize=12, fontweight='bold')
    ax.set_xlabel(condition.replace("_", " ").title())
    ax.set_ylabel('Average Students')
    ax.set_xticklabels(['No', 'Yes'], rotation=0)
    ax.grid(axis='y', alpha=0.3)
    
    # Add insight text
    normal = condition_stats[False]
    special = condition_stats[True]
    diff_pct = ((special - normal) / normal * 100)
    ax.text(0.5, 0.95, f'Impact: {diff_pct:+.1f}%', 
            transform=ax.transAxes, ha='center', va='top',
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.tight_layout()
plt.show()

print("\nüí° INSIGHTS:")
for condition in conditions:
    stats = df.groupby(condition)['students_attended'].mean()
    normal = stats[False]
    special = stats[True]
    diff = special - normal
    diff_pct = (diff / normal * 100)
    print(f"   - {condition}: {diff_pct:+.1f}% change ({diff:+.0f} students)")
```

```python
# ============================================
# CELL 11: FOOD WASTE ANALYSIS
# ============================================

print("="*60)
print("FOOD WASTE ANALYSIS")
print("="*60)

print(f"Total Food Prepared: {df['food_prepared_kg'].sum():,.2f} kg")
print(f"Total Food Consumed: {df['food_consumed_kg'].sum():,.2f} kg")
print(f"Total Food Wasted: {df['food_wasted_kg'].sum():,.2f} kg")
print(f"Average Waste per Meal: {df['food_wasted_kg'].mean():.2f} kg")
print(f"Average Waste Percentage: {df['waste_percentage'].mean():.2f}%")

print(f"\nüí∞ COST ANALYSIS:")
print(f"Total Cost: ‚Çπ{df['total_cost_rs'].sum():,.2f}")
print(f"Total Waste Cost: ‚Çπ{df['waste_cost_rs'].sum():,.2f}")
daily_waste_cost = df.groupby('date')['waste_cost_rs'].sum().mean()
print(f"Average Daily Waste Cost: ‚Çπ{daily_waste_cost:,.2f}")
print(f"Potential Monthly Savings: ‚Çπ{daily_waste_cost * 30:,.2f}")

# Visualization
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
df['date'] = pd.to_datetime(df['date'])
daily_waste = df.groupby('date')['food_wasted_kg'].sum()
plt.plot(daily_waste.index, daily_waste.values, color='red', alpha=0.7, linewidth=1.5)
plt.title('Daily Food Waste Over Time', fontsize=14, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Food Wasted (kg)')
plt.xticks(rotation=45)
plt.grid(alpha=0.3)

plt.subplot(2, 2, 2)
waste_by_meal = df.groupby('meal_type')['food_wasted_kg'].mean()
waste_by_meal.plot(kind='bar', color=['#FF6B6B', '#4ECDC4', '#45B7D1'], edgecolor='black')
plt.title('Average Waste by Meal Type', fontsize=14, fontweight='bold')
plt.xlabel('Meal Type')
plt.ylabel('Average Waste (kg)')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)

plt.subplot(2, 2, 3)
waste_by_day = df.groupby('day_of_week')['food_wasted_kg'].mean().reindex(day_order)
waste_by_day.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Average Waste by Day of Week', fontsize=14, fontweight='bold')
plt.xlabel('Day')
plt.ylabel('Average Waste (kg)')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)

plt.subplot(2, 2, 4)
menu_waste = df.groupby('menu_item')['food_wasted_kg'].mean().sort_values(ascending=False).head(10)
plt.barh(range(len(menu_waste)), menu_waste.values, color='darkred', alpha=0.7)
plt.yticks(range(len(menu_waste)), menu_waste.index)
plt.xlabel('Average Waste (kg)')
plt.title('Top 10 Menu Items by Waste', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()
plt.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()
```
